```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import os
```

```python
def weather_and_load(b_ids):
  missing_ids = []
  for b_id in b_ids:
    FOLDER_PATH = PATH + "/building_data/" + b_id
    if (os.path.exists(FOLDER_PATH)):
      weather = pd.read_csv(FOLDER_PATH + "/weather.csv")
      load = pd.read_csv(FOLDER_PATH + "/load.csv")
      print(weather.head())
      print(load.head())
    else:
      missing_ids.append(b_id)

  print("These ids were missing from the dataset:", missing_ids)
```

```python
weather_and_load(["332130", "2500", "222"])
```

```python
df = pd.read_csv(PATH + "/building_data/" + "332130" +"/weather.csv")

# Convert 'date_time' column to datetime
df['date_time'] = pd.to_datetime(df['date_time'])

# Create time series plot with Plotly
fig = px.line(df, x='date_time', y=['Dry Bulb Temperature [°C]', 'Relative Humidity [%]'],
              labels={'value': 'Measurement', 'date_time': 'Time'},
              title="Time Series of Temperature and Humidity")

# Show plot
fig.show()
```
