```python
import pandas as pd
import os
import re
import plotly.express as px
```

```python
!ls '/content/drive/My Drive/Team-Fermata-Energy/[EXTERNAL] breakthrough_tech_ai_f24/data/'
PATH = "/content/drive/My Drive/Team-Fermata-Energy/[EXTERNAL] breakthrough_tech_ai_f24/data"
```

```python
df = pd.read_csv(PATH + '/metadata.csv')
df.head()
```

```python
filtered_df = df[(df['in.state'] == 'CA') & (df['in.vintage'] == '1990 to 1999')]

ca_vintage_bldg_ids = filtered_df['bldg_id']

print(ca_vintage_bldg_ids)
```

```python
building_folders_path = '/content/drive/My Drive/Team-Fermata-Energy/[EXTERNAL] breakthrough_tech_ai_f24/data/building_data/'
```

```python
building_folders = [f for f in os.listdir(building_folders_path) if os.path.isdir(os.path.join(building_folders_path, f))]
```

```python
print(building_folders)
```

```python
def clean_name(name):
    return re.sub(r'\s+', '', str(name).strip()).lower()
```

```python
ca_vintage_bldg_ids_clean = set(clean_name(bldg_id) for bldg_id in ca_vintage_bldg_ids)

valid_vintage_building_folders = [bldg_id for bldg_id in building_folders if clean_name(bldg_id) in ca_vintage_bldg_ids_clean]

print("Valid Building Folders (vintage '1990 to 1999'):")
print(valid_vintage_building_folders)
print(len(valid_vintage_building_folders))
```

```python
for bldg_id in valid_vintage_building_folders[:10]:
    file_path = os.path.join(building_folders_path, bldg_id, 'load.csv')

    if os.path.exists(file_path):
        df = pd.read_csv(file_path)

        if 'timestamp' in df.columns and 'out.electricity.total.energy_consumption' in df.columns:
            df['timestamp'] = pd.to_datetime(df['timestamp'])
            df['out.electricity.total.energy_consumption'] = pd.to_numeric(df['out.electricity.total.energy_consumption'], errors='coerce')

            fig = px.line(df, x='timestamp', y='out.electricity.total.energy_consumption', title=f'Electricity Usage for Building {bldg_id}')

            fig.show()
```
